# 사진 불러오기

```python
import numpy as np
import matplotlib.pyplot as plt
```

### 이미지 불러오기

```python
bob = plt.imread('bob_dylan.jpg')
```

```python
plt.imshow(bob)
```

```python
bob.shape # 행렬 형태
```

```python
bob.dtype # 숫자 타입
```

```python
bob.max(), bob.min() # 최대, 최소값
```

- ravel() : 다차원 행렬을 쭉 펼쳐서 일차원 목록 형태로 만든다
- np.bincount() : 0 부터 최대값 까지의 값이 가지는 갯수를 출력한다

```python
np.bincount(bob.ravel())
```

```python
plt.hist(bob.ravel(), bins=100, color='gray') # 히스토그램
```

```python
plt.hist(bob[:,:,0].ravel(), bins=100, color='red') # Red 채널의 히스토그램
pass
```

```python
plt.hist(bob[:,:,1].ravel(), bins=100, color='green') # Green 채널의 히스토그램
pass
```

```python
plt.hist(bob[:,:,2].ravel(), bins=100, color='blue') # Blue 채널의 히스토그램
pass
```

```python
plt.figure(figsize=[10,6])

plt.subplot(1,3,1)
plt.hist(bob[:,:,0].ravel(), bins=100, color='red')
plt.title('Red channel')

plt.subplot(1,3,2)
plt.hist(bob[:,:,1].ravel(), bins=100, color='green')
plt.title('Green channel')

plt.subplot(1,3,3)
plt.hist(bob[:,:,2].ravel(), bins=100, color='blue')
plt.title('Blue channel')
```

### 채널 분리

```python
r = bob[:, :, 0]
g = bob[:, :, 1]
b = bob[:, :, 2]

r.shape
```

```python
plt.imshow(r, vmin=0, vmax=255, cmap='Reds')
plt.colorbar()
```

```python
plt.imshow(g, vmin=0, vmax=255, cmap='Greens')
plt.colorbar()
```

```python
plt.imshow(b, vmin=0, vmax=255, cmap='Blues')
plt.colorbar()
```

### 채널 이미지 합치기

```python
bob2 = np.zeros(bob.shape, dtype='uint8') # 0~255 값을 가지므로, 'uint8' 타입으로 지정해야 한다
bob2[:,:,0] = r
bob2[:,:,1] = g
bob2[:,:,2] = b
plt.imshow(bob2)
```

- 채널 바꾸기

```python
bob2 = np.zeros(bob.shape, dtype='uint8')
bob2[:,:,0] = b
bob2[:,:,1] = g
bob2[:,:,2] = r
plt.imshow(bob2)
```

- 채널 빼기

```python
bob2 = np.zeros(bob.shape, dtype='uint8')
#bob2[:,:,0] = r
bob2[:,:,1] = g
bob2[:,:,2] = b
plt.imshow(bob2)
```

### 사진 잘라내기

```python
bob2 = bob[:200, :200]
bob2.shape
```

```python
plt.imshow(bob2)
```

- 잘라낸 이미지 붙이기

```python
bob3 = bob.copy()
bob3[-200:, -200:] = bob2

plt.imshow(bob3)
```

### 이미지 저장

```python
plt.imsave('bob3.jpg', bob3)
```

- 저장한 이미지를 'irfanview' 이미지뷰어를 이용해 불러와 보자.

```python
plt.imshow(lucy)
plt.colorbar()
```
